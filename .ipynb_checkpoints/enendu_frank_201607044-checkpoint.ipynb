#### A. Logistics Regression Model

##### i. Build Model

Baseline

```python
# This code computes the baseline accuracy for a classification problem.
acc_baseline = y_train.value_counts(normalize=True).max()
print("Baseline Accuracy:", round(acc_baseline, 2))
```

The baseline accuracy of a model is 0.65. This means that the model would be correct 65% of the time by simply guessing the majority class. In other words, any model that we develop should perform better than the baseline accuracy of 0.65 to be considered useful.

Iterate

```python
# Initialising model
model = LogisticRegression(random_state=42, solver='saga', max_iter=10000)
```

```python
# define the parameter grid
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'penalty': ['l1', 'l2']}
```

```python
# Perform K-Fold Cross Validation with k = 5 using GridSearchCV
cross_val = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
cross_val.fit(X_train, y_train)
```

```python
# Build model model
model = make_pipeline(
    LogisticRegression(max_iter=10000,
                       C=cross_val.best_params_['C'], 
                       penalty=cross_val.best_params_['penalty'], 
                       random_state=42,
                       solver='saga'
 )
)
# Fit model to training data
model.fit(X_train, y_train)
```

Evaluate

```python
# Evaluating the model base on the best parameter obtained from the cross validation
acc_train = accuracy_score(y_train, model.predict(X_train))
# acc_val = model.score(X_val, y_val)

print("Training Accuracy:", round(acc_train, 2))
# print("Validation Accuracy:", round(acc_val, 2))
```

The training accuracy of a model is 0.97, indicating that the model has accurately classified 97% of the training data. However, it is important to evaluate the model's performance on an independent test set to ensure it can generalize well to unseen data.

```python
# View the result of the model prediction on the X_train classes
model.predict(X_train)[:5]
```

```python
# Checking Confidence level on the Train Dataset
y_train_pred_proba = model.predict_proba(X_train)
y_train_pred_proba[:5]
```

The output shows the predicted probability of the target variable for the first five instances in the train dataset using the trained model. The first column represents the probability of the instance belonging to class 0, while the second column represents the probability of the instance belonging to class 1.

```python
# Testing Model on the test dataset
model_pred = model.predict(X_train)
```

```python
# Printing Confusion Metrix for the test
confusion_metrix = confusion_matrix(y_train, model_pred)
confusion_metrix
```

#### B. Decision Tree Model

##### i. Build Model

Train, Test, Validation Split

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

We will adopt a different approach for hyperparameter tuning in this decision tree model. Instead of using cross-validation, we will use the train-validation-test split.

Iterate

We trained the model with a maximum of 7 tree depth, but later determined the optimal depth to be 4.

```python
# Build Decision Tree Model
model = make_pipeline(
    DecisionTreeClassifier(max_depth=4, random_state=42)
)
# Fit model to training data
model.fit(X_train, y_train)
```

```python
# Calculating accuracy of model for train dataset
acc_train = accuracy_score(y_train, model.predict(X_train))
# acc_val = model.score(X_val, y_val)

print("Training Accuracy:", round(acc_train, 2))
# print("Validation Accuracy:", round(acc_val, 2))
```

```python
tree_depth = model.named_steps["decisiontreeclassifier"].get_depth()
print("Tree Depth:", tree_depth)
```

```python
depth_hyperparams = range(1, 10, 1)
```

```python
# Create empty lists for training and validation accuracy scores
training_acc = []
validation_acc = []

for d in depth_hyperparams:
    # Create model with `max_depth` of `d`
    test_model = make_pipeline(
        DecisionTreeClassifier(max_depth=d, random_state=42)
    )
    # Fit model to training data
    test_model.fit(X_train, y_train)
    # Calculate training accuracy score and append to `training_acc`
    training_acc.append(test_model.score(X_train, y_train))
    # Calculate validation accuracy score and append to `training_acc`
    validation_acc.append(test_model.score(X_val, y_val))

print("Training Accuracy Scores:", training_acc[:3])
print("Validation Accuracy Scores:", validation_acc[:3])
```

The output shows the training and validation accuracy scores for a model over three different iterations. The training accuracy scores are 0.9308, 0.9643, and 0.9643, while the validation accuracy scores are 0.9018, 0.9286, and 0.9286. The model appears to perform well on both the training and validation sets, with a slight decrease in performance from training to validation sets.

```python
# Plot `depth_hyperparams`, `training_acc`
plt.plot(depth_hyperparams, training_acc, label="Training")
plt.plot(depth_hyperparams, validation_acc, label="Validation")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy Score")
plt.legend();
```

We plotted the train and validation accuracy against the depth of the decision tree and observed that initially, both accuracies increased together. However, at a tree depth of 4, the validation accuracy began to decrease while the training accuracy continued to increase. This indicates that the model started to overfit on the training data beyond a tree depth of 4.

#### B. Random Forest

##### i. Build Model

Resample

below, we oversamples the training data using the RandomOverSampler algorithm to handle class imbalance. This means that instances from the minority class are randomly sampled with replacement to create a balanced dataset. The new oversampled dataset is stored in the X_train_over and y_train_over variables. The fit_resample method of the RandomOverSampler class is called to perform the oversampling, and the shape of the resulting X_train_over dataset is printed. Finally, the first few rows of the oversampled dataset are displayed using the head() method.

```python
# Here we resample our dataset with replacement
over_sampler = RandomOverSampler(random_state=42)
X_train_over, y_train_over = over_sampler.fit_resample(X_train, y_train)
print("X_train_over shape:", X_train_over.shape)
X_train_over.head()
```

Notice that we used the RandomOverSampler because are classes are imballance

Iterate

```python
# Build our model
model = make_pipeline(SimpleImputer(), RandomForestClassifier(random_state=42))
print(model)
```

```python
# Calculate modelaccuracy on the 
model_acc_scores = cross_val_score(model, X_train_over, y_train_over, cv=5, n_jobs=-1)
print(model_acc_scores)
```

The output is an array of the accuracy scores for each of the 5 folds. The model had high accuracy scores across all folds, ranging from 0.965 to 0.986, indicating good performance and consistency.

```python
params = {
    "simpleimputer__strategy": ["mean", "median"],
    "randomforestclassifier__n_estimators": range(25, 100, 25), 
    "randomforestclassifier__max_depth": range(10, 50, 10)
}
params
```

```python
model = GridSearchCV(
    model,
    param_grid=params,
    cv=5,
    n_jobs=-1,
    verbose=1
)
model
```

The output shows the result of a GridSearchCV object, which is a cross-validation method to optimize hyperparameters of a machine learning model. The GridSearchCV has been performed on a Pipeline object, which includes a SimpleImputer and a RandomForestClassifier model with a random state of 42. The GridSearchCV was performed with 5-fold cross-validation, and the hyperparameters that were searched are the imputation strategy for SimpleImputer, the number of estimators for RandomForestClassifier, and the maximum depth for RandomForestClassifier. The output also shows that the verbose parameter was set to 1 to provide more details during the training process, and n_jobs was set to -1 to use all available CPU cores.

```python
# Train model
model.fit(X_train_over, y_train_over)
```

The code shows the GridSearchCV function being called with 24 candidate parameter settings, 5-fold cross-validation and parallel computation using all available CPU cores. The parameter grid being searched has three hyperparameters: 'simpleimputer__strategy', 'randomforestclassifier__n_estimators', and 'randomforestclassifier__max_depth', with the specified ranges. The output shows the estimator used, the parameter grid, and other specifications.

```python
cv_results = pd.DataFrame(model.cv_results_)
cv_results.head()
```

The output of cv_results.head() displays the first five rows of the DataFrame cv_results, which includes columns for hyperparameters, mean test score, standard deviation of test scores, and other cross-validation statistics. The possible output will depend on the number of hyperparameters tuned, the range of values for each hyperparameter, and the cross-validation strategy used.

```python
cv_results["param_randomforestclassifier__max_depth"] == 10
```

We then plot the mean fit time of the Random Forest classifier model against the number of estimators used, with a fixed maximum depth of 10.

The plot shows how increasing the number of estimators affects the training time of the model. The x-axis represents the number of estimators used, and the y-axis represents the mean fit time in seconds.

```python
# Create mask
mask = cv_results["param_randomforestclassifier__n_estimators"] == 25
# Plot fit time vs max_depth
plt.figure(figsize = (4,3))
plt.plot(
    cv_results[mask]["param_randomforestclassifier__max_depth"],
    cv_results[mask]["mean_fit_time"]
)
# Label axes
plt.xlabel("Max Depth")
plt.ylabel("Mean Fit Time [seconds]")
plt.title("Training Time vs Max Depth (n_estimators=25)");
```

The plot shows the mean fit time increasing with the maximum depth, which suggests that increasing the maximum depth leads to a more complex model that takes longer to train.
